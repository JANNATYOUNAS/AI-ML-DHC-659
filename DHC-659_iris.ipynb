```python
#  Exploring and Visualizing a Simple Dataset
# Importing Libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Desktop\\Iris.csv")
```

```python
df
```

```python
# Basic info
print("Shape of dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Summary statistics
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Scatter Plot
sns.pairplot(df, hue='Species')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
# Histograms
df.hist(bins=20, figsize=(10, 6), edgecolor='black')
plt.suptitle("Histograms of Features", y=1.02)
plt.show()
```

```python
# Box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plots of Iris Features")
plt.show()
```
